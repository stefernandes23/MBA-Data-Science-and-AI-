# Exercícios - Listas

### 4. Faça um Programa que leia um vetor de 5 números inteiros e mostre-os.

```python
# lista vazia
num = []

# solicitando 5 números inteiros
for i in range (5):
    numero = int(input(f"Insira o número {i + 1}: "))
    num.append(numero)

# retorna os números inseridos
print("Os números inseridos foram: ")
for numero in num:
    print(numero)
```

### 5.  Faça um Programa que leia 20 números inteiros e armazene-os num vetor. Armazene os números pares no vetor PAR e os números ímpares no vetor IMPARES. Imprima os vetores.

```python
# lista vazia
num_par = []
num_impar = []

# solicitando 20 números inteiros
for i in range (20):
    num = int(input(f"Insira o número {i + 1}: "))

    # verificando se é par ou ímpar
    if num % 2 == 0:
        num_par.append(num)
    else: num_impar.append(num)
    
# par ou ímpar?
print("Números pares: ", num_par)
print("Números ímpares: ", num_impar)
```

### 6. Faça um Programa que leia um vetor de 10 caracteres, e diga quantas consoantes foram lidas. Imprima as consoantes.

```python
# lista vazia
caracteres = []
consoantes = []

#vogais
vogais = ['a', 'e', 'i', 'o', 'u']

# solicitando 10 caracteres
for i in range (10):
    caractere = input (f"Insira o caractere {i + 1}: ")
    caracteres.append(caractere)
    
# o caractere é uma consoante?
for caractere in caracteres:
    if caractere.isalpha() and caractere not in vogais:
        consoantes.append(caractere)
        
# quantas consoantes foram lidas?
qtd_consoantes = len(consoantes)

#retorna consoantes e a quantidade
print ("Consoantes: ", consoantes)
print("Quantidade: ", qtd_consoantes)
```

### 8. Dada a tupla ponto = (3, 4), calcule a distância do ponto à origem (0, 0) usando a fórmula matemática:   
$\sqrt{(xˆ2+yˆ2)}$
, onde: x e y são as coordenadas do ponto.

```python
import math

#tupla ponto
tupla_ponto = (3,4)

#coordenadas
x,y = tupla_ponto

#cálculo distância ponto à origem
distancia = math.sqrt(x**2 + y**2)

#distância
print(f"A distância do ponto ({x}, {y}) à origem (0, 0) é {distancia}.")
```

### 11. Crie um dicionário chamado agenda com nomes de pessoas e seus números de telefone. Permita que o usuário adicione novos contatos à agenda.  

```python
# dicionário agenda
agenda = {}

# adicionar contatos
while True:
    print("\nMenu de opções: ")
    print ("1. Adicionar novo contato")
    print("2. Mostrar a agenda")
    print("3. Sair")
    
    opcao = input("Escolha uma opção: ").strip()
    
    if opcao == '1': 
        nome = input("Digite o nome do contato: ").strip()
        cel = input ("Digite o número de celular: ").strip()
        agenda[nome] = cel
        print(f"{nome} adicionado(a) com sucesso :) ")
    elif opcao == '2': 
        if agenda:
            print ("\nAgenda de contatos ")
            for nome, cel in agenda.items():
                print(f"{nome}: {cel}")
        else:
            print("\nA agenda está vazia.")
    elif opcao == '3':
        print("Você saiu.")
        break
    else:
        print("Opção inválida. Tente novamente.")
```
